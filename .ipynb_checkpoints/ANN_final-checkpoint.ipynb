```python
class ANN():
    def __init__(self, neurons_hidden_layers, input_shape, output_shape = 1):
        num_hidden_layers = len(neurons_hidden_layers)
        self.total_layers = int(num_hidden_layers + 2)
        self.shape_layers = [None]*self.total_layers
        self.shape_layers[0],self.shape_layers[-1] = input_shape, output_shape
        
        for i in range(num_hidden_layers):
            self.shape_layers[i+1] = neurons_hidden_layers[i]
        self.weights = {}
        for i in range(1, len(self.shape_layers)):
            self.weights[i] = np.array(np.random.uniform(0,1,(self.shape_layers[i-1],self.shape_layers[i])))

    def sigmoid(self, output):
        return 1/(1 + np.exp(-output))

    def dsigmoid(self, output):
        return self.sigmoid(output)*(1 -  self.sigmoid(output))
    
    def loss(self, activation, y): 
        return -y*np.log(activation) - (1 - y)*np.log(1 - activation)

    def dloss(self, activation,y):
        return (activation - y)/(activation*(1-activation))
    
    def feed_forward(self, input_sample):
        output = {}
        activation = {}
        output[0] = input_sample
        activation[0] = input_sample
        for i in range(1, len(self.weights)+1):
            output[i] = self.weights[i].T.dot(activation[i-1])
            activation[i] = self.sigmoid(output[i])
        return output, activation

    
    def backpropagate_error(self,activation, output, eta, y):
        delta = {}
        delta[0] = 0
        for i in reversed(range(1,len(activation))):
            if i==len(activation)-1:
                delta[i] = np.multiply(self.dloss(activation[i],y), self.dsigmoid(output[i])) 
            else:
                delta[i] = np.multiply(self.weights[i+1].dot(delta[i+1]) , self.dsigmoid(output[i]))
            d = np.array(delta[i]).reshape(1,-1)
            a = np.array(activation[i-1]).reshape(-1,1)
            gradient = a.dot(d)
            self.weights[i] = self.weights[i] - eta*gradient

    def train_weights(self, X_train, y_train, eta, epochs):
        cost_hist = []
        acc_hist = []
        for epoch in range(epochs+1):
            cost = 0
            count = 0
            for i,sample in enumerate(X_train):
                output, activation = self.feed_forward(sample)
                loss_sample = self.loss(activation[len(self.weights)][0],y[i])
                cost += loss_sample
                self.backpropagate_error(activation, output, eta, y[i])
                prediction = round(activation[len(self.weights)][0])
                count += 1 if prediction==y[i] else 0
            accuracy = count/len(X_train)
            cost_hist.append(cost)
            acc_hist.append(accuracy)

            if epoch%200 == 0:
                print(f'epoch: {epoch} \t cost: {cost}\t acc: {accuracy}')

            if accuracy >= 0.99:
                print(f'Early Stop \nepoch: {epoch} \t cost: {cost} \t acc: {count/len(X_train)}')
                return self.weights, cost_hist,acc_hist
                
        return self.weights,cost_hist, acc_hist
    
    def predict(self, X_test):
        prediction_output = []
        for i,sample in enumerate(X_test):
            output, activation = self.feed_forward(sample)
            prediction = round(activation[len(self.weights)][0])
            prediction_output.append(prediction)
        return prediction_output
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons, make_blobs
import pandas as pd
import numpy as np
np.seterr(divide='ignore', invalid='ignore')
#np.random.seed(10)


X,y = make_moons(n_samples=100 ,noise=0.2)
df = pd.DataFrame(dict(x1=X[:,0], x2=X[:,1], label=y))
fig, ax = plt.subplots()
ax.scatter(df[df.label == 1].x1, df[df.label==1].x2, c='b', label='1')
ax.scatter(df[df.label == 0].x1, df[df.label == 0].x2, c= 'r', label='0')
ax.legend()
ax.set_title('Binary Classification Example', fontsize=20)
ax.set_xlabel(r'$W_1$',fontsize=20)
ax.set_ylabel(r'$W_2$',fontsize=20)
plt.show()
fig.savefig('moon_dataset.png')
```

## Test1: 0 hidden layers

```python
neurons_hidden_layers = []
input_shape = 2
output_shape = 1
eta = 0.1
epochs =  2000

network = ANN( neurons_hidden_layers, input_shape)
weights,cost_hist,acc_hist = network.train_weights(X, y, eta, epochs)
```

```python
from matplotlib.colors import ListedColormap

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

h = 0.01
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))


Z = np.array( network.predict( np.c_[xx.ravel(), yy.ravel()]) )
Z = Z.reshape(xx.shape) 

cmap_light = ListedColormap(['#FFAAAA', '#AAAAFF'])
plt.pcolormesh(xx, yy, Z, cmap=cmap_light)
plt.scatter(df[df.label == 1].x1, df[df.label==1].x2, c='blue', label='1')
plt.scatter(df[df.label == 0].x1, df[df.label == 0].x2, c= 'red', label='0')
plt.legend()
plt.text(1.7, -1.4, f'Accuracy: {acc_hist[-1]}', color='k', 
        bbox=dict(facecolor='white', alpha=0.7, edgecolor='gray', pad=10.0))


plt.xlabel(r'$W_1$', fontsize=20)
plt.ylabel(r'$W_2$', fontsize=20)
plt.title(' 0 Hidden Layers ', fontsize=20)
plt.savefig('test1.png')
```
